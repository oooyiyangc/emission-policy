```python
import numpy as np
import pandas as pd
```

```python
cc = pd.read_csv("../data/cleaned-data/country_continent.csv")
cc
```

```python
pop = pd.read_csv("../data/cleaned-data/pop.csv")[["name", "code", "2010"]]
pop
```

```python
cc_pop = cc.merge(pop[["code", "2010"]], left_on="Three_Letter_Country_Code", right_on="code")
cc_pop = cc_pop.rename(columns={"2010":"pop2010"})
cc_pop = cc_pop[["Continent_Name", "Continent_Code", "Country_Name", "Three_Letter_Country_Code", "pop2010"]]
cc_pop
```

```python
cc_pop.to_csv("../data/cleaned-data/cc.csv", index=False)
```
